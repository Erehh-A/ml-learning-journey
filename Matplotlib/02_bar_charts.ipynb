### Bar Charts and Analyzing Data from CSVs


```python
import pandas as pd
from matplotlib import pyplot as plt 
from collections import Counter

plt.style.use('fivethirtyeight')

df = pd.read_csv('data_for_bars.csv')
# print(df)

ids = df['Responder_id']
langs = df['LanguagesWorkedWith']

langs_counter = Counter()
for lag in langs:
    langs_counter.update(lag.split(';'))

languages = []
popularity = []

for item in langs_counter.most_common(20):
    languages.append(item[0])
    popularity.append(item[1])

languages.reverse()           # Coz it was showing upside down
popularity.reverse()

plt.barh(languages, popularity)

plt.title('Languages V/S People using them')
plt.xlabel('Number of people using them')
plt.ylabel('Languages')

plt.show()
```
