```python
from matplotlib import pyplot as plt
import pandas as pd

plt.style.use('seaborn-v0_8-whitegrid')

df = pd.read_csv('data_for_scatter_plots.csv')
# print(df)
views = df['view_count']
likes = df['likes']
ratios = df['ratio']

plt.scatter(views, likes, c=ratios, cmap='summer', edgecolors='black')

cbar = plt.colorbar()
cbar.set_label('Ratios of Likes v/s Dislikes')

plt.xscale('log')
plt.yscale('log')

plt.title('Views v/s Likes in Trending videos')
plt.xlabel('Views (Log)')
plt.ylabel('Likes (Log)')

plt.tight_layout()
plt.show()
```
