Accessing/Changing elements, rows and columns

```python
import numpy as np
a = np.array([[6, 7, 2, 8, 1, 87], [12, 65, 12, 67, 87, 45]])

#a[row, column] for the element
num = a[1, 1]
print(num)
print(a[1, :])
```

Linear Algebra

```python
# Matrix Multiplication
import numpy as np
a = np.ones((4, 5))
print(a)
b = np.full((5, 10), 2)
print(b)

c = np.matmul(a, b)
print(c)
```
