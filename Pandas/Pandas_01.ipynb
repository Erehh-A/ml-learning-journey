Making changes to data

```python
import pandas as pd
df = pd.read_csv('pokemon_data.txt', sep="\t")
# print(df)

# Making a new column "Total" of the sum of stats of each pokimon
df['Total'] = df.iloc[:, 4:10].sum(axis = 1)  # Axis = 0 (vertical), axis = 1 (Horizontal)
# print(df)

# Moving that newly added column
cols = list(df.columns)
df = df[cols[0 : 4] + [cols[-1]] + cols[4 : 12]]
# print(df)
```

Saving the data

```python
import pandas as pd
df = pd.read_csv('pokemon_data.txt', sep="\t")
df['Total'] = df.iloc[:, 4:10].sum(axis=1)
col = list(df.columns)
df = df[cols[0:4]+[col[-1]]+col[4:12]]
# df.to_csv('modified.csv', index=False)

#df.to_excel('modified.xlsx', index=False)

df.to_csv('modified.txt', index=False, sep='\t')
```
